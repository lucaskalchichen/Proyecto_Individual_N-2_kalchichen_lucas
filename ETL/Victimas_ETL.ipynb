    Descartaremos columnas inutimes

* **AAAA** ,**MM**, **DD** por ser datos de la columna fecha 

```python
drop_columns = [  'AAAA', 'MM', 'DD']

df_victimas.drop(columns=drop_columns, inplace=True)
```

Visualizamos y Normalizamos


```python
#edad
df_victimas['EDAD'].unique()
```

```python
# Convert the 'edad' column to numeric values with NaN for non-numeric values
df_victimas['EDAD'] = pd.to_numeric(df_victimas['EDAD'], errors='coerce')
```

```python
#fecha
df_victimas['FECHA_FALLECIMIENTO']
```

```python
# Convertir la columna 'FECHA_FALLECIMIENTO' a tipo datetime
df_victimas['FECHA_FALLECIMIENTO'] = pd.to_datetime(df_victimas['FECHA_FALLECIMIENTO'], errors='coerce')

# Extraer la fecha sin la hora y convertirla a tipo date
df_victimas['FECHA_FALLECIMIENTO'] = df_victimas['FECHA_FALLECIMIENTO'].dt.date

```

```python
df_victimas['ROL'].unique()
```

```python
df_victimas['VICTIMA'].unique()
```

```python
df_victimas['SEXO'].unique()
```

```python
# Lista de columnas tipo texto
cols = ['ROL', 'VICTIMA', 'SEXO']

# Aplicar cambios a las columnas seleccionadas
for col in cols:
    df_victimas[col] = df_victimas[col].str.strip()
```

Renombramo columnas

```python
Actuales = df_victimas.columns
Nuevas = ['id_hecho', 'fecha','rol', 'vehiculo_victima', 'sexo', 'edad', 'fecha_falleciomiento']

columnas = {Actuales: Nuevas for (Actuales, Nuevas) in zip(Actuales, Nuevas)}
df_victimas.rename(columns=columnas, inplace=True)
```
