Normalizacion de Datos

```python
#hora
df_hechos['HH']
```

```python

# Reemplazar 'sd' por NaN en la columna 'HH'
df_hechos['HH'].replace('sd', np.nan, inplace=True)

# Convertir la columna a valores numéricos enteros
df_hechos['HH'] = pd.to_numeric(df_hechos['HH'], errors='coerce').astype('Int64')
```

```python

# Lista de columnas tipo texto
cols = ['LUGAR_DEL_HECHO', 'TIPO_DE_CALLE', 'VICTIMA', 'ACUSADO']

# Aplicar cambios a las columnas seleccionadas
for col in cols:
    df_hechos[col] = df_hechos[col].str.strip()
```

```python
#comuna
df_hechos['COMUNA'].unique()
#comuna va del 1 al 15 porque son las 15 comunas de la CABA los 0 seran los nulos selos remplasa Nan numpy
```

```python

df_hechos['COMUNA'].replace(0, np.nan, inplace=True)
df_hechos['COMUNA']
```

Renombro Colunmbas 

```python
Actuales = df_hechos.columns
Nuevas = ['id_hecho', 'cantidad_victimas', 'fecha', 'hora', 'lugar_del_hecho', 
       'tipo_de_calle', 'comuna', 'longitud', 'latitud', 'vehiculo_victima', 'vehiculo_acusado']

columnas = {Actuales: Nuevas for (Actuales, Nuevas) in zip(Actuales, Nuevas)}
df_hechos.rename(columns=columnas, inplace=True)
```
